```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials
import pandas as pd
from sqlalchemy import create_engine
import yaml
import time
import sys
#!
with open("volunteers.yml", "r", encoding="utf-8") as file:
    config = yaml.safe_load(file)

# Пример доступа к параметрам:
engine = config["database"]["connection"]["url"]
schema = config["database"]["schema"]
pg_table_name = config["database"]["table"]
sheet_url = config["paths"]["sheet_url"]
cr_path = config["paths"]["credentials"]

# Настройка подключения к Google Sheets
def connect_to_google_sheets():
    # Укажите путь к вашему JSON-файлу с учетными данными
    scope = ['https://spreadsheets.google.com/feeds',
             'https://www.googleapis.com/auth/drive']
    
    credentials = ServiceAccountCredentials.from_json_keyfile_name(
        'oncologicags-59363649872f.json', scope)
        #{cr_path}, scope)
    
    gc = gspread.authorize(credentials)
    return gc

# Чтение данных из Google Sheets
def read_sheet_data(gc, url, sheet_name, range_name):
    try:
        # Открываем таблицу по URL
        spreadsheet = gc.open_by_url(url)
        
        # Выбираем конкретный лист
        worksheet = spreadsheet.worksheet(sheet_name)
        
        # Получаем данные из указанного диапазона
        data = worksheet.get(range_name)
        
        return data
    except Exception as e:
        print(f"Ошибка при чтении данных из Google Sheets: {e}")
        return None

# URL таблицы Google Sheets Сотрудники/Волонтеры
#sheet_url = "https://docs.google.com/spreadsheets/d/1hyBbj_bDZgKBxKhsApbNHmWH0jnC2xq_r7tT5VMprUI/"
     
# Имя таблицы в PostgreSQL
#pg_table_name = "stg_volunteers_by_direction"   

# Подключение к PostgreSQL
#engine = create_engine("postgresql+psycopg2://postgres:pgSwordFish13#@localhost:5432/postgres")

# Имя листа и диапазон данных (например, "A1:D10")
sheet_name = "Псих. направление"
range_name = "D2:F1000"

# Подключаемся к Google Sheets
gc = connect_to_google_sheets()
    
 # Читаем данные
data = read_sheet_data(gc, sheet_url, sheet_name, range_name)

# Пример DataFrame
df = pd.DataFrame(data, columns=["employee", "department", "job_title"])

# Запись в таблицу с указанием схемы
df.to_sql(
    pg_table_name,
    engine,
    schema="employees",  # Указываем схему
    if_exists="append",  # 'append' - добавить, 'replace' - перезаписать
    index=False
)
print("Данные успешно загружены в PostgreSQL через Pandas!")

time.sleep(5)

# Имя листа и диапазон данных (например, "A1:D10")
sheet_name = "Равное конс."
range_name = "D2:F1000"

# Подключаемся к Google Sheets
gc = connect_to_google_sheets()
    
 # Читаем данные
data = read_sheet_data(gc, sheet_url, sheet_name, range_name)

# Пример DataFrame
df = pd.DataFrame(data, columns=["employee", "department", "job_title"])

# Запись в таблицу с указанием схемы
df.to_sql(
    pg_table_name,
    engine,
    schema="employees",  # Указываем схему
    if_exists="append",  # 'append' - добавить, 'replace' - перезаписать
    index=False
)
print("Данные успешно загружены в PostgreSQL через Pandas!")

time.sleep(5)

# Имя листа и диапазон данных (например, "A1:D10")
sheet_name = "Команда СЗМ"
range_name = "D2:F1000"

# Подключаемся к Google Sheets
gc = connect_to_google_sheets()
    
 # Читаем данные
data = read_sheet_data(gc, sheet_url, sheet_name, range_name)

# Пример DataFrame
df = pd.DataFrame(data, columns=["employee", "department", "job_title"])

# Запись в таблицу с указанием схемы
df.to_sql(
    pg_table_name,
    engine,
    schema="employees",  # Указываем схему
    if_exists="append",  # 'append' - добавить, 'replace' - перезаписать
    index=False
)
print("Данные успешно загружены в PostgreSQL через Pandas!")

time.sleep(5)

# Имя листа и диапазон данных (например, "A1:D10")
sheet_name = "Юр. помощь"
range_name = "D2:F1000"

# Подключаемся к Google Sheets
gc = connect_to_google_sheets()
    
 # Читаем данные
data = read_sheet_data(gc, sheet_url, sheet_name, range_name)

# Пример DataFrame
df = pd.DataFrame(data, columns=["employee", "department", "job_title"])

# Запись в таблицу с указанием схемы
df.to_sql(
    pg_table_name,
    engine,
    schema="employees",  # Указываем схему
    if_exists="append",  # 'append' - добавить, 'replace' - перезаписать
    index=False
)
print("Данные успешно загружены в PostgreSQL через Pandas!")

time.sleep(5)

# Имя листа и диапазон данных (например, "A1:D10")
sheet_name = "Команда ГЛ"
range_name = "D2:F1000"

# Подключаемся к Google Sheets
gc = connect_to_google_sheets()
    
 # Читаем данные
data = read_sheet_data(gc, sheet_url, sheet_name, range_name)

if not data:
    df = pd.DataFrame(columns=["employee", "department", "job_title"])  # Создаём пустой DF с колонками
else:
    df = pd.DataFrame(data[1:], columns=data[0])

# Пример DataFrame
#df = pd.DataFrame(data, columns=["employee", "department", "job_title"])
if df.empty:
    #print("DataFrame пуст. Таблица не будет создана/обновлена.")
    time.sleep(0.5)
else:    
# Запись в таблицу с указанием схемы
    df.to_sql(
        pg_table_name,
        engine,
        schema="employees",  # Указываем схему
        if_exists="append",  # 'append' - добавить, 'replace' - перезаписать
        index=False,
        dtype={"employee": String, "department": String, "job_title": String}
    )
    
#print(data)
print("Данные успешно загружены в PostgreSQL через Pandas!")
```
